```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "StudyHours": [3,4,2,5,6,3,4,5,6,2,4,3,5,6,4,3,5,6,4,2],
    "SleepHours": [7,6,8,7,6,7,8,6,7,8,7,6,7,8,6,7,8,6,7,8],
    "ScreenTime": [5,6,4,5,7,6,5,4,6,5,7,6,5,4,6,7,5,6,4,5],
    "SportsHours": [1,2,1,2,3,1,2,3,2,1,2,3,1,2,3,2,1,2,3,1],
    "GPA": [7.8,8.2,7.5,8.5,8.7,7.9,8.1,8.3,8.6,7.7,8.0,7.8,8.4,8.8,8.1,7.9,8.5,8.7,8.2,7.6]
}

df = pd.DataFrame(data)

df
```

```python
print("Mean:\n", df.mean())

print("\nMedian:\n", df.median())

print("\nStandard Deviation:\n", df.std())

print("\nCorrelation Matrix:\n", df.corr())
```

Scatter plot

```python
plt.figure()

plt.scatter(df["StudyHours"], df["GPA"])

plt.xlabel("Study Hours")
plt.ylabel("GPA")
plt.title("Study Hours vs GPA")

plt.show()
```

Histogram

```python
plt.figure()

plt.hist(df["SleepHours"], bins=5)

plt.xlabel("Sleep Hours")
plt.ylabel("Frequency")
plt.title("Distribution of Sleep Hours")

plt.show()
```

line plot

```python
plt.figure()

plt.plot(df["SportsHours"])

plt.xlabel("Student Index")
plt.ylabel("Sports Hours")
plt.title("Sports Activity Trend")

# plt.show()
plt.plot(df["SportsHours"])

plt.xlabel("Student Index")
plt.ylabel("Sports Hours")
plt.title("Sports Activity Trend")

plt.show()
```
