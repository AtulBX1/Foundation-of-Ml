1. Why are the values of Marks in float?
The list contains integer values (85, 88, 90...), but also None values.

The reason: Pandas uses NumPy under the hood. NumPy's integer dtype (int64) cannot represent NaN/None — it has no concept of a "missing integer". So when Pandas encounters None in an integer list, it automatically upcasts the entire column to float64, because float64 can represent NaN (as numpy.nan).
