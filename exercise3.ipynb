```python
import numpy as np
```

```python
values = np.random.randint(120,650,size=(4,7))
print(values)
```

```python
print(values.shape,values.ndim,values.size,values.dtype)
```

```python
coloumn_mean = np.mean(values, axis=1)
row_means = np.mean(values, axis=0)
```

```python
status = np.max(coloumn_mean)
status2 = np.max(row_means)
print(status)
print(f"{status2:.2f}")
```

```python
status = np.where(values>550,values*1.20,values)
print(status)
```

```python
status = np.where(values>row_means,True,False)
print(status)
```
