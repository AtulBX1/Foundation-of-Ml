DataFrames in Pandas

```python
data = {"Name":["Aman","Krish","Rohan"],"Age":[22,24,26],"Marks":[85,90,88]}
```

```python
print(data)
type(data)
```

```python
data_df = pd.DataFrame(data,index = ["Student1","Student2","student3"])
print(data_df)
type(data_df)
```

```python
data_df.set_index("Name",inplace=True)
# note: if we are setting inplace = true it means we are changing original dataset itselt
print(data_df)
```
