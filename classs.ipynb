Where in Numpy

```python
arr = np.array([1,-3,20,-5,15])
result = np.where(arr<0,0,arr)
print(result)
result = np.where(arr %2 != 0, -1, arr)
print(result)
```

```python
arr = np.array([50,45,67,89,39])
status = np.where(arr >=50,1,0)
print(status)
```

```python
data  = np.array([10,20,30,40,50])
mean_val = np.mean(data)
result = np.where(data > mean_val, mean_val,data)
print(result)
```

```python
arr = np.array([[10,60,30],[80,20,50]])
result = np.where(arr>=50,1,0)
print(result)
```
