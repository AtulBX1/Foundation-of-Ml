Data Visulization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Name":["Aman","Piya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age":[22,25,24,23,26,27,28,22],
    "Gender":["Male","Female","Male","Female","Male","Female","Male","Female"],
    "Marks":[85,76,95,56,82,97,65,75]
}
df = pd.DataFrame(data)
```

```python
plt.plot(df["Marks"]) #univarient 
plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(df["Name"],df["Marks"]) # bivarient analysis
plt.title("Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(df["Age"],bins = 5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df["Age"],df["Marks"])
plt.title("Age vs MArks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

Gender Distribution

```python
gender_counts = df["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index,autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

Customizing plots
example with grids and colors

```python
plt.bar(df["Name"],df["Marks"],color= "purple")
plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(df["Age"])
plt.title("Age Distribution")

plt.subplot(1,2,2)
plt.bar(df["Name"],df["Marks"])
plt.title("Marks of Students")

plt.show()
```
